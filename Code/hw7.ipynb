## Question 4

```python
# Initial conditions
r = 0.35 # decay rate
initial_impurity = 405.0 # parts per gallon
target = 0.6
intervals = initial_impurity
values = [initial_impurity]
while intervals >= target:
    intervals *= r
    values.append(intervals)
applications = len(values) - 1
x = range(applications + 1)
y = values
plt.figure(figsize=(10,10))
plt.plot(x, y)
for i in zip(x, y):
    plt.annotate('(%.2f, %.2f)' % i, xy=i)
plt.grid()
plt.xlabel('Applications')
plt.ylabel('Parts Per Gallon')
print(f"It takes {applications} applications to make the water drinkable")
```
