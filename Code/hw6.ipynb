## Question 3

### 3.a

```python
start_t = 0
end_t = 100
dt = 1
t = np.arange(start_t, end_t, dt)
p0 = 110000
k = 0.02
K = 500000
def equation(p, t):
    dpdt = k * p
    dpdt *= 1 - (p/K)
    return dpdt
ans_p = odeint(equation, p0, t)
```

### 3.b

```python
fig, ax = plt.subplots()
ax.plot(t, ans_p)
ax.grid()
plt.xlabel('Time (t)')
plt.ylabel('Population (P)')
ax.set_xlim([0, end_t])
```

### 3.c

```python
analyticalP = K/(1 + ((K - p0)/p0) * np.exp(-k * t))
fig, ax = plt.subplots()
ax.plot(t, analyticalP, 'r', t, ans_p, 'k*', markevery=3)
ax.grid()
ax.legend(['Analytical', 'Scipy'])
ax.set_xlim([0, end_t])
```
