## Question 1

```python
def pick():
    ans = np.array([np.random.randint(1,46)])
    c = 0
    while c < 4:
        to_add = np.random.randint(1,46)
        if to_add not in ans:
            ans = np.append(ans, to_add)
            c += 1    
    ans = np.append(ans, np.random.randint(1,10))    
    return ans
pick()
```

## Question 3

```python
def roll():
    first_dice = np.random.randint(1,7)
    second_dice = np.random.randint(1,7)
    dice_sum = first_dice + second_dice
    if dice_sum % 2 == 0:
        print(f"{first_dice}\n{second_dice}\nSum of two dices is even")
    else:
        print(f"{first_dice}\n{second_dice}\nSum of two dices is odd")
roll()
```

## Question 4

```python
def soccer():
    outcomes = ['Home Team Wins', 'Draw', 'Away Team Wins']
    
    # Team A vs Team B
    game_1_p = [0.35, 0.35, 0.30]
    game_1 = np.random.choice([0,1,2], p=game_1_p)

    # Team C vs Team D
    game_2_p = [0.30, 0.20, 0.50]
    game_2 = np.random.choice([0,1,2], p=game_2_p)
    
    # Team E vs Team F
    game_3_p = [0.85, 0.10, 0.05]
    game_3 = np.random.choice([0,1,2], p=game_3_p)
    
    print(f"Game 1: {outcomes[game_1]}\nGame 2: {outcomes[game_2]}\nGame 3: {outcomes[game_3]}")
soccer()
```
